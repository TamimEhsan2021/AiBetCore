```python
def reducer(link:str):
    split_link = link[24:].split('/')
    prelink = '/'.join(split_link[:-1])
    full_link = prelink+"/"+split_link[-1][:4]
    return(full_link)
```

```python
df["reduced_link"] = df["Link"].apply(reducer)
```

```python
def eval_result(result:str):
    result_list = result.split("-")
    try:
        if int(result_list[0]) > int(result_list[1]):
            return "Home Win"
        elif int(result_list[0]) < int(result_list[1]):
            return "Away Win"
        else:
            return "Draw"
    except:
        return "Unknown"
```
